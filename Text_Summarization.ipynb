## Pegasus Model

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model1 = AutoModelForSeq2SeqLM.from_pretrained('google/pegasus-xsum')
tokenizer1 = AutoTokenizer.from_pretrained('google/pegasus-xsum')
```

```python
TokensInputPeg = tokenizer1.encode("summarize: "+ data, return_tensors='pt', max_length=512, truncation=True)
ids = model1.generate(TokensInputPeg, min_length=80, max_length=120, no_repeat_ngram_size=1)
PegSummary = tokenizer1.decode(ids[0], skip_special_tokens=True)
```

```python
PegSummary
```

```python
with open('PegsusSummary.txt', 'w+') as f:
    for items in PegSummary:
        f.write('%s' %items)
    print("File written successfully")
f.close()
```

## Bart Model

```python
tokenizer2 = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model2 = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")
```

```python
TokensInputBart = tokenizer2.encode("summarize: "+ data, return_tensors='pt', max_length=512, truncation=True)
ids2 = model2.generate(TokensInputBart, min_length=80, max_length=120)
BartSummary = tokenizer2.decode(ids2[0], skip_special_tokens=True)
```

```python
BartSummary
```

```python
with open('BartSummary.txt', 'w+') as f:
    for items in BartSummary:
        f.write('%s' %items)
    print("File written successfully")
f.close()
```
